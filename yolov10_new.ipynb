#**Roboflow website potential**

**1. Object Detection:**
  This project is used to identify objects and their positions with bounding boxes.
  This is best for:
    1. Counting objects (Objects in conveyor)
    2. Tracking (moving objects)

**2. Classification:**
  This tool box assign labels to the entire image
  Classification Type are:
    1. Multi-Label  
    2. Single-Label

  This tool is best for
    1. Filtering objects
    2. Content Moderation

**3. Instance Segmentation:**
  Detect multiple objects and their actual shape
  This tool is best for
    1. Taking measurements
    2. Odd shapes

**4. Keypoint Detection:**
  This tool is used for identifying keypoints ("skeletons") to subjects
  This tool is best
    1. Pose estimation (human/animal)

**5. Semantic Segmentation:**
  This tool is used for assigning every pixel to a lable. Often inferior to instance segmentation
  This tool is best for:
    1. Robotics
    2. Autonomous Driving

```python
import supervision as sv
```

```python
from ultralytics import YOLOv10

model = YOLOv10(f'/content/gdrive/MyDrive/yolov_10_new/runs/detect/train/weights/best.pt')

dataset = sv.DetectionDataset.from_yolo(
    images_directory_path=f"/content/gdrive/MyDrive/yolov_10_new/brain_tumor_new/valid/images",
    annotations_directory_path=f"/content/gdrive/MyDrive/yolov_10_new/brain_tumor_new/valid/labels",
    data_yaml_path=f"/content/gdrive/MyDrive/yolov_10_new/data.yml"
)

bounding_box_annotator = sv.BoundingBoxAnnotator()
label_annotator = sv.LabelAnnotator()
```

```python
import random

random_image = random.choice(list(dataset.images.keys()))
random_image = dataset.images[random_image]

results = model(source=random_image, conf=0.25)[0]
detections = sv.Detections.from_ultralytics(results)

annotated_image = bounding_box_annotator.annotate(
    scene=random_image, detections=detections)
annotated_image = label_annotator.annotate(
    scene=annotated_image, detections=detections)

sv.plot_image(annotated_image)
```
